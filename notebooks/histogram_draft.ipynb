```python
import cv2
from matplotlib import pyplot as plt
from matplotlib import ticker

def show_image_histogram_2d(image, bins=24, tick_spacing=5):
    fig, axes = plt.subplots(1, 3, figsize=(12, 5))
    channels_mapping = {0: 'B', 1: 'G', 2: 'R'}
    for i, channels in enumerate([[0, 1], [0, 2], [1, 2]]):
        hist = cv2.calcHist(
            [image], channels, None, [bins] * 2, [1, 256] * 2)
#         hist = remove_b(hist, 5)
        channel_x = channels_mapping[channels[0]]
        channel_y = channels_mapping[channels[1]]

        ax = axes[i]
        ax.set_xlim([0, bins - 1])
        ax.set_ylim([0, bins - 1])

        ax.set_xlabel(f'Channel {channel_x}')
        ax.set_ylabel(f'Channel {channel_y}')
        ax.set_title(f'2D Color Histogram for {channel_x} and '
                     f'{channel_y}')

        ax.yaxis.set_major_locator(
            ticker.MultipleLocator(tick_spacing))
        ax.xaxis.set_major_locator(
            ticker.MultipleLocator(tick_spacing))

        im = ax.imshow(hist)

    fig.colorbar(im, ax=axes.ravel().tolist(),
                 orientation='orizontal')
    fig.suptitle(f'2D Color Histograms with {bins} bins',
                 fontsize=16)
    plt.show()
```

```python
import numpy as np # linear algebra
import json
from matplotlib import pyplot as plt
from skimage import color
from skimage.feature import hog
from sklearn import svm
from sklearn.metrics import classification_report,accuracy_score

import pathlib
import pandas as pd
import os
```

```python
import tqdm
```

```python
def hog_feature(image, ppc):
    fd, hog_image = hog(image, orientations=8, pixels_per_cell=(ppc, ppc),
                    cells_per_block=(1, 1), visualize=True, multichannel=True)
    return fd, hog_image
```

```python
ppc = 16
hog_images = []
hog_features = []
labels = []
image_path = []
for image in tqdm.tqdm(pathlib.Path(dirr).rglob('*.jp*')):
    img = cv2.imread(str(image))
    fd, hog_image = hog_feature(img, ppc)
#     fd, hog_image = hog(img, orientations=8, pixels_per_cell=(ppc, ppc),
#                     cells_per_block=(1, 1), visualize=True, multichannel=True)
    label = image.parent.stem
    if label == '.ipynb_checkpoints': 
        continue
    hog_features.append(fd)
    hog_images.append(hog_image)
    labels.append(label)
    image_path.append(image)
```

```python
labels = np.array(labels).reshape(-1,1)
dataframe = np.hstack((hog_features,labels))
np.random.shuffle(dataframe)
```

```python
percentage = 80
partition = int(len(hog_features)*percentage/100)
x_train, x_test = dataframe[:partition,:-1],  dataframe[partition:,:-1]
y_train, y_test = dataframe[:partition,-1:].ravel() , dataframe[partition:,-1:].ravel()
```

```python
clf = svm.SVC(C=1.0)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
print("Accuracy: "+str(accuracy_score(y_test, y_pred)))
print('\n')
print(classification_report(y_test, y_pred))
```

## HISTOGRAM VIS 

```python
from PIL import Image

import matplotlib.pyplot as plt

 
```

```python
def getRed(redVal):

    return '#%02x%02x%02x' % (redVal, 0, 0)

 

def getGreen(greenVal):

    return '#%02x%02x%02x' % (0, greenVal, 0)

   

def getBlue(blueVal):

    return '#%02x%02x%02x' % (0, 0, blueVal)

 

 

# Create an Image with specific RGB value

image = Image.open(DIR + '/205.jpg')
# image = cv2.imread()
 
```

```python
# Modify the color of two pixels

image.putpixel((0,1), (1,1,5))

image.putpixel((0,2), (2,1,5))

 

# Display the image

image.show()
```

```python
# Get the color histogram of the image

histogram = image.histogram()
```

```python
 

# Take only the Red counts

l1 = histogram[1:256]

 

# Take only the Blue counts

l2 = histogram[257:512]

 

# Take only the Green counts

l3 = histogram[513:768]

 

plt.figure(0)

 

# R histogram

for i in range(0, 255):

    plt.bar(i, l1[i], color = getRed(i), edgecolor=getRed(i), alpha=0.3)

 

# G histogram

plt.figure(1)

for i in range(0, 255):

    plt.bar(i, l2[i], color = getGreen(i), edgecolor=getGreen(i),alpha=0.3)

 

 

# B histogram

plt.figure(2)

for i in range(0, 255):

    plt.bar(i, l3[i], color = getBlue(i), edgecolor=getBlue(i),alpha=0.3)

 

plt.show()
```
