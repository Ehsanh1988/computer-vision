```python
def p2f(input_farsi):
    b=''
    for char in input_farsi:
        try:
            b += finglish[char]
        except:
#             print(char)
            b += char
    return '_'.join(b.split(' '))

allprods = allprods.assign(secCat_finglish = allprods['SecCat'].apply(p2f))
```
